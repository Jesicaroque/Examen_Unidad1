```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dfam = pd.read_csv('amazon.csv').astype(str)
dfap = pd.read_csv('apple.csv').astype(str)
dfbl = pd.read_csv('blockchain.csv').astype(str)
dfbm = pd.read_csv('bmw.csv').astype(str)
dfco = pd.read_csv('coinbase.csv').astype(str)
dfde = pd.read_csv('dell.csv').astype(str)
dffa = pd.read_csv('facebook.csv').astype(str)
dfgo = pd.read_csv('google.csv').astype(str)
dfhu = pd.read_csv('huawei.csv').astype(str)
dfle = pd.read_csv('lenovo.csv').astype(str)
dfmi = pd.read_csv('microsoft.csv').astype(str)
dfni = pd.read_csv('nissan.csv').astype(str)
dfpi = pd.read_csv('pinteres.csv').astype(str)
dfse = pd.read_csv('seika.csv').astype(str)
dfte = pd.read_csv('tesla.csv').astype(str)
dftr = pd.read_csv('trust.csv').astype(str)
dftw = pd.read_csv('twitter.csv').astype(str)
dfub = pd.read_csv('uber.csv').astype(str)




def formatito():
    dfam['Date'] = pd.to_datetime(dfam['Date'], format="%b %d, %Y")
    dfap['Date'] = pd.to_datetime(dfap['Date'], format="%b %d, %Y")
    dfbl['Date'] = pd.to_datetime(dfbl['Date'], format="%b %d, %Y")
    dfbm['Date'] = pd.to_datetime(dfbm['Date'], format="%b %d, %Y")
    dfco['Date'] = pd.to_datetime(dfco['Date'], format="%b %d, %Y")
    dfde['Date'] = pd.to_datetime(dfde['Date'], format="%b %d, %Y")
    dffa['Date'] = pd.to_datetime(dffa['Date'], format="%b %d, %Y")
    dfgo['Date'] = pd.to_datetime(dfgo['Date'], format="%b %d, %Y")
    dfhu['Date'] = pd.to_datetime(dfhu['Date'], format="%b %d, %Y")
    dfle['Date'] = pd.to_datetime(dfle['Date'], format="%b %d, %Y")
    dfmi['Date'] = pd.to_datetime(dfmi['Date'], format="%b %d, %Y")
    dfni['Date'] = pd.to_datetime(dfni['Date'], format="%b %d, %Y")
    dfpi['Date'] = pd.to_datetime(dfpi['Date'], format="%b %d, %Y")
    dfse['Date'] = pd.to_datetime(dfse['Date'], format="%b %d, %Y")
    dfte['Date'] = pd.to_datetime(dfte['Date'], format="%b %d, %Y")
    dftr['Date'] = pd.to_datetime(dftr['Date'], format="%b %d, %Y")
    dftw['Date'] = pd.to_datetime(dftw['Date'], format="%b %d, %Y")
    dfub['Date'] = pd.to_datetime(dfub['Date'], format="%b %d, %Y")
    
    
    dfam['Close']= dfam['Close'].str.replace(',', '')
    dfam['Close'] = pd.to_numeric(dfam['Close'], errors = 'coerce')
    dfap['Close'] = pd.to_numeric(dfap['Close'], errors = 'coerce')
    dfbl['Close'] = pd.to_numeric(dfbl['Close'], errors = 'coerce')
    dfbm['Close'] = pd.to_numeric(dfbm['Close'], errors = 'coerce')
    dfco['Close'] = pd.to_numeric(dfco['Close'], errors = 'coerce')
    dfde['Close'] = pd.to_numeric(dfde['Close'], errors = 'coerce')
    dffa['Close'] = pd.to_numeric(dffa['Close'], errors = 'coerce')
    dfgo['Close'] = pd.to_numeric(dfgo['Close'], errors = 'coerce')
    dfhu['Close']= dfhu['Close'].str.replace(',', '')
    dfhu['Close'] = pd.to_numeric(dfhu['Close'], errors = 'coerce')
    dfle['Close'] = pd.to_numeric(dfle['Close'], errors = 'coerce')
    dfmi['Close'] = pd.to_numeric(dfmi['Close'], errors = 'coerce')
    dfni['Close'] = pd.to_numeric(dfni['Close'], errors = 'coerce')
    dfpi['Close'] = pd.to_numeric(dfpi['Close'], errors = 'coerce')
    dfse['Close'] = pd.to_numeric(dfse['Close'], errors = 'coerce')
    dfte['Close'] = pd.to_numeric(dfte['Close'], errors = 'coerce')
    dftr['Close'] = pd.to_numeric(dftr['Close'], errors = 'coerce')
    dftw['Close'] = pd.to_numeric(dftw['Close'], errors = 'coerce')
    dfub['Close'] = pd.to_numeric(dfub['Close'], errors = 'coerce')
    
    
def comportamiento():
    grafica(dfam.sort_values(by = ['Date']), "Amazon")
    grafica(dfap.sort_values(by = ['Date']), "Apple")
    grafica(dfbl.sort_values(by = ['Date']), "Blockchain")
    grafica(dfbm.sort_values(by = ['Date']), "BMW")
    grafica(dfco.sort_values(by = ['Date']), "Coinbase")
    grafica(dfde.sort_values(by = ['Date']), "Dell")
    grafica(dffa.sort_values(by = ['Date']), "Facebook")
    grafica(dfgo.sort_values(by = ['Date']), "Google")
    grafica(dfhu.sort_values(by = ['Date']), "Huawei")    
    grafica(dfle.sort_values(by = ['Date']), "Lenovo")
    grafica(dfmi.sort_values(by = ['Date']), "Microsoft")
    grafica(dfni.sort_values(by = ['Date']), "Nissan")
    grafica(dfpi.sort_values(by = ['Date']), "Pinterest")
    grafica(dfse.sort_values(by = ['Date']), "Seika")
    grafica(dfte.sort_values(by = ['Date']), "Tesla")
    grafica(dftr.sort_values(by = ['Date']), "Trust Wallet")
    grafica(dftw.sort_values(by = ['Date']), "Twitter")
    grafica(dfub.sort_values(by = ['Date']), "Uber")

def grafica(df, name):
    plt.figure(figsize=(20, 5))
    plt.plot(df['Date'],df['Close'])
    plt.xlabel("Año")
    plt.ylabel("Precio de la acción")
    plt.title("{}".format(name))
    plt.show()
    
formatito()
comportamiento()
```
