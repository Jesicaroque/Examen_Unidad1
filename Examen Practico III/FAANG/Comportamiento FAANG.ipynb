```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dffa = pd.read_csv('stock-facebook.csv').astype(str)
dfam = pd.read_csv('stock-amazon.csv').astype(str)
dfap = pd.read_csv('stock-apple.csv').astype(str)
dfne = pd.read_csv('stock-netflix.csv').astype(str)
dfgo = pd.read_csv('stock-google.csv').astype(str)

def formatito():
    dffa['Date'] = pd.to_datetime(dffa['Date'], format="%b %d, %Y")
    dfam['Date'] = pd.to_datetime(dfam['Date'], format="%b %d, %Y")
    dfap['Date'] = pd.to_datetime(dfap['Date'], format="%b %d, %Y")
    dfne['Date'] = pd.to_datetime(dfne['Date'], format="%b %d, %Y")
    dfgo['Date'] = pd.to_datetime(dfgo['Date'], format="%b %d, %Y")
    
    dffa['Close'] = pd.to_numeric(dffa['Close'], errors = 'coerce')
    dfam['Close']= dfam['Close'].str.replace(',', '')
    dfam['Close'] = pd.to_numeric(dfam['Close'], errors = 'coerce')
    dfap['Close'] = pd.to_numeric(dfap['Close'], errors = 'coerce')
    dfne['Close'] = pd.to_numeric(dfne['Close'], errors = 'coerce')
    dfgo['Close'] = pd.to_numeric(dfgo['Close'], errors = 'coerce')
    
def comportamiento():
    grafica(dffa.sort_values(by = ['Date']), "Facebook", "9", "b")
    grafica(dfam.sort_values(by = ['Date']), "Amazon", "10", "y")
    grafica(dfap.sort_values(by = ['Date']), "Apple", "10", "m")
    grafica(dfne.sort_values(by = ['Date']), "Netflix", "10", "r")
    grafica(dfgo.sort_values(by = ['Date']), "Google", "10", "g")

def grafica(df, name, years, color):
    plt.figure(figsize=(20, 8))
    plt.plot(df['Date'],df['Close'], color)
    plt.xlabel("Año")
    plt.ylabel("Precio de la acción")
    plt.title("{} desde hace {} años".format(name, years))
    plt.show()
    
formatito()
comportamiento()
```
